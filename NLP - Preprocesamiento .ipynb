En esta etapa vamos a proceder a crear una funcion para preprocesar las reviews contenidas en nuestro dataset. Incluiremos en esta funcion tokenizacion, lematizacion, etiquetado gramatical, asi como todas las tecnicas requeridas por las instrucciones de la practica, como eliminar stopwords, o eliminar signos de puntuacion.

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
# Lematizador y stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Funcion de preprocesado completa

def preprocess_text(text):
    # Minusculas y limpieza basica
    text = re.sub(r"[^\w\s]", "", str(text).lower())

    # Tokenizacion
    tokens = word_tokenize(text)

    # Etiquetado gramatical
    tagged_tokens =pos_tag(tokens)

    # Lematizacion
    lemmatized = [
        lemmatizer.lemmatize(word, get_wordnet_pos(pos))
        for word, pos in tagged_tokens
        if word not in stop_words and word.isalpha()
    ]

    return " ".join(lemmatized)
```

Ahora que la funcion esta definida, vamos a cargar el dataset y aplicarle la funcion.

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP/Practica/reviews_amazon_books_balanced.csv')
```

```python
df['clean_text'] = df['reviewText'].apply(preprocess_text)
```
